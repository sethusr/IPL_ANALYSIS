#Does winning the toss has any advantage? Lets check


```python
#Analysing the relationshop between the toss winner and winning team
#Interating over each row of DF to count the number of times the team that won the toss also won the match
#Initializing the variables to 0 before the loop
win_count = 0
lost_count = 0
#Checking if the toss winner and winning team have the same value
for index, value in matches_df.iterrows():
#If yes, variable 'win_count' is incremented
    if(value['TossWinner']==value['WinningTeam']):
#         print(value['winner'])
        win_count += 1
#If no, then variable 'loss_count' is incremented
    elif (value['TossWinner'] != value['WinningTeam']):
        lost_count +=1

print(f'The number of times the team winning toss have won: {win_count}')
print(f'The number of times the team winning toss have lost: {lost_count}')
prob = win_count/len(matches_df)
print(f'The probability of winning if won the toss: {prob:.2f}')    

```

Its almost 50 - 50. 
While winning the toss does provide a slight edge with a 1% higher chance of winning, it would be naive to consider it a significant advantage. In fact, there were 461 instances where the team that lost the toss still managed to win the game.

## Run distribution Per Year

```python
#Converting the 'Date' column into a datetime format
matches_df['Season'] = pd.to_datetime(matches_df['Date']).dt.year
```

```python
#Creating a DF to merge the 'matches_df' DF on the 'ID' column with the deliveries
batsmen = matches_df[['ID','Season']].merge(deliveries, left_on = 'ID', right_on = 'ID', how = 'left').drop('ID', axis = 1)
#Grouping the rows in the DF by 'Season'
Season=batsmen.groupby(['Season'])['total_run'].sum().reset_index()
```

```python
#Creating a new DF by grouping 'batsmen' DF by 'Season and calculating the number of sixes hit in each season
Season_boundaries=batsmen.groupby("Season")["batsman_run"].agg(lambda x: (x==6).sum()).reset_index()
fours=batsmen.groupby("Season")["batsman_run"].agg(lambda x: (x==4).sum()).reset_index()
Season_boundaries=Season_boundaries.merge(fours,left_on='Season',right_on='Season',how='left')
Season_boundaries=Season_boundaries.rename(columns={'batsman_run_x':'6"s','batsman_run_y':'4"s'})
```

```python
#Multipying the number of sixes and fours hit in each season by '6' and '4'
Season_boundaries['6"s'] = Season_boundaries['6"s']*6
Season_boundaries['4"s'] = Season_boundaries['4"s']*4
Season_boundaries['total_run'] = Season['total_run']
```

```python
#Creating a stacked bar chart that shows the distribution of runs scored per year in the IPL
trace1 = go.Bar(
    x=Season_boundaries['Season'],
    y=Season_boundaries['total_run']-(Season_boundaries['6"s']+Season_boundaries['4"s']),
    marker = dict(line=dict(color='#000000', width=1)),
    name='Remaining runs',opacity=0.6)

trace2 = go.Bar(
    x=Season_boundaries['Season'],
    y=Season_boundaries['4"s'],
    marker = dict(line=dict(color='#000000', width=1)),
    name='Run by 4"s',opacity=0.7)

trace3 = go.Bar(
    x=Season_boundaries['Season'],
    y=Season_boundaries['6"s'],
    marker = dict(line=dict(color='#000000', width=1)),
    name='Run by 6"s',opacity=0.7)


data = [trace1, trace2, trace3]
layout = go.Layout(title="Run Distribution per year",barmode='stack',xaxis = dict(tickmode='linear',title="Year"),
                                    yaxis = dict(title= "Run Distribution"), plot_bgcolor='rgb(245,245,245)')

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

Just slight increase in runs by boundaries over years

# Bowler Analysis

```python
#Groups the deliveries by bowler
#Calculates their total runs given, number of balls bowled, and the number of overs bowled
bowlers=deliveries.groupby('bowler').sum().reset_index()
bowl=deliveries['bowler'].value_counts().reset_index()
bowlers=bowlers.merge(bowl,left_on='bowler',right_on='index',how='left')
bowlers=bowlers[['bowler_x','total_run','bowler_y']]
bowlers.rename({'bowler_x':'bowler','total_run':'runs_given','bowler_y':'balls'},axis=1,inplace=True)
bowlers['overs']=(bowlers['balls']//6)
```

```python
#Creating a new column 'bowlers to the DF which counts the number of wickets taken by each bowler
dismissal_kinds = ["bowled", "caught", "lbw", "stumped", "caught and bowled", "hit wicket"]  
ct=deliveries[deliveries["kind"].isin(dismissal_kinds)]
ct=ct['bowler'].value_counts().reset_index()
bowlers=bowlers.merge(ct,left_on='bowler',right_on='index',how='left').dropna()
bowlers=bowlers[['bowler_x','runs_given','overs','bowler_y']]
bowlers.rename({'bowler_x':'bowler','bowler_y':'wickets'},axis=1,inplace=True)
bowlers['economy']=(bowlers['runs_given']/bowlers['overs'])
```

# Batsman Analysis

```python
#Used to compute various batting statistics for each batsmen in the IPL
df_strike_rate = deliveries.groupby(['batter']).agg({'ballnumber':'count','batsman_run':'mean'}).sort_values(by='batsman_run',ascending=False)
df_strike_rate.rename(columns ={'batsman_run' : 'strike rate'}, inplace=True)
df_runs_per_match = deliveries.groupby(['batter','ID']).agg({'batsman_run':'sum'})
df_total_runs = df_runs_per_match.groupby(['batter']).agg({'sum' ,'mean','count'})
df_total_runs.rename(columns ={'sum' : 'batsman_run','count' : 'match count','mean' :'average score'}, inplace=True)
df_total_runs.columns = df_total_runs.columns.droplevel()
df_sixes = deliveries[['batter','batsman_run']][deliveries.batsman_run==6].groupby(['batter']).agg({'batsman_run':'count'})
df_four = deliveries[['batter','batsman_run']][deliveries.batsman_run==4].groupby(['batter']).agg({'batsman_run':'count'})
'''The resulting DF contains the stike rate, total runs scored, 
average score per match, number of sixes, and number of fours for 
each batsmen in the IPL'''
df_batsman_stat = pd.merge(pd.merge(pd.merge(df_strike_rate,df_total_runs, left_index=True, right_index=True),
                                    df_sixes, left_index=True, right_index=True),df_four, left_index=True, right_index=True)
```

```python
df_batsman_stat.rename(columns = {'ballnumber' : 'ballnumber', 'strike rate':'strike_rate','batsman_run' : 'batsman_run',
                                  'match count' : 'match_count','average score' : 'average_score' ,'batsman_run_x' :'six',
                                  'batsman_run_y':'four'},inplace=True)
df_batsman_stat['strike_rate'] = df_batsman_stat['strike_rate']*100
df_batsman_stat.sort_values(by='batsman_run',ascending=False,inplace=True)
#df_batsman_stat.sort_values(by='batsman_run',ascending=False)
df_batsman_stat.reset_index(inplace=True)
```
